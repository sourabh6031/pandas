### MERGING

```python
users = pd.DataFrame({'userid':[1,2,3],'name':['sharad','shahid','kushali']})
```

```python
msgs = pd.DataFrame({'userid':[1,1,2,4],'msg':['hmm','acha','thik hai','nice']})
```

```python
users
```

```python
msgs
```

```python
# inner join
users.merge(msgs,on='userid',how='inner')
```

```python
# outer join
users.merge(msgs,on='userid',how='outer')
```

```python
users
```

```python
msgs
```

```python
# left join
users.merge(msgs,on='userid',how='left')
```

```python
# right join
users.merge(msgs,on='userid',how='right')
```

```python
#if we have tables with different column name  
```

```python
users.rename(columns = {'userid':'id'},inplace=True)
users
```

```python
# use left_on and right_on 
users.merge(msgs,left_on='id',right_on = 'userid',how='right')
```

```python
users.merge(msgs,left_on='id',right_on = 'userid',how='outer')
```
