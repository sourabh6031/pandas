### LOC and ILOC

```python
# loc = explicit indexing - custom made indexing
# iloc = implicit indexing - automatically defined (based on 0 based indexing)
```

```python
mckinsey.loc[[2]]
```

```python
mckinsey.iloc[[2]]
```

```python
# since index starts with 0, [:3] = [0,1,2]. value will show till 3rd index but with explicitely defined index; here 1,2,3 
mckinsey.iloc[:3]
```

```python
# loc uses explicitely defined index here (1,2,3). [:3] and it also include last position
mckinsey.loc[:3]
```

```python
# selecting multiple rows
mckinsey.loc[[2,4,5]]
```
