to calculate temperature mean for each drug temperature_avg as new column 

```python
data_tidy
```

```python
data_tidy.groupby('Drug_Name')['Temperature'].mean()
```

```python
data_tidy.groupby(['Drug_Name','Date'])['Temperature'].mean()
```

create a new column 'avg_temp' with mean value of respective drug.

```python
# apply function expects original dataframe in return

def temp_avg(df):
    df['temperature_avg'] = df['Temperature'].mean()
    return df
```

```python
# so this will create a new column with mean value for each Drug_Name across all date 
data_tidy = data_tidy.groupby(['Drug_Name']).apply(temp_avg)
```

```python
data_tidy
```

```python
# this will replace all NaN value with given column i.e. temperature_avg 

data_tidy['Temperature'].fillna(data_tidy['temperature_avg'],inplace=True)
```

```python
# to check NaN value replaced or not
np.array(data_tidy['Temperature'])
```

```python
# so this will create a new column with mean value for each Drug_Name and for each date 
data_tidy.groupby(['Drug_Name','Date']).apply(temp_avg)
```

```python
# like temperature we will do same for pressure

def pressure_avg(df):
    df['pressure_avg'] = df['Pressure'].mean()
    return df
```

```python
data_tidy = data_tidy.groupby('Drug_Name').apply(pressure_avg)
```

```python
data_tidy
```

```python
data_tidy['Pressure'] = data_tidy['Pressure'].fillna(data_tidy['pressure_avg'])
```

```python
data_tidy
```

```python
# this is same as code in line 113.Here we directly passing column in the function.
def avg_pressure(x):
    data_tidy['avg_pres'] = x.mean()
    return x
```

```python
data_tidy.groupby('Drug_Name')['Pressure'].mean()
```

```python
np.array(data_tidy.groupby('Drug_Name')['Pressure'].apply(avg_pressure))
```
