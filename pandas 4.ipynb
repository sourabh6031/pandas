```python
# correct solution
```

```python
# we define a function which takes dataframe as a parameter,
# then store avg_rev of revenue column
# then we compared budget to avg_rev and we stored in new column 'risky'
# lastly we return the dataframe

# this function need to pass in apply() with groupby function, 
# so that avg_rev is calculated for each director and then can be compared. 
```

```python
def risky_movie(df):
    avg_revenue = df['revenue'].mean()
    df['risky'] = df['budget'] > avg_revenue
    return df
```

```python
data = data.groupby('director_name').apply(risky_movie)
```

```python
data
```

```python
data['risky'].value_counts()
```

### MULTIPLE INDEXING

#### MULTIPLE  AGGREGATIONS ON MULTIPLE COLUMNS

```python
data.groupby('director_name')[['year','title']].aggregate(['min','max'])
```

```python
data_agg = data.groupby('director_name')[['year','title']].aggregate({'year':['min','max'],'title':['count']})
```

```python
data_agg
```

```python
# we have year and title as first level index and min,max,count as second level index
data_agg.columns
```

```python
# to reset as year_min, year_max and title_count we can use join method.
```

```python
# 1st approach
```

```python
data_agg.columns
```

```python
data_agg.columns = ['_'.join(i) for i in data_agg.columns]
```

```python
data_agg.columns
```

```python
data_agg = data_agg.reset_index()
```

```python
# 2nd approach
```

```python
# in aggregate function we create a new column name without string as type and pass (column,aggregation) as tuple.
data_agg = data.groupby('director_name')[['year','title']].aggregate(year_max=('year','max'),year_min=('year','min'),title_count=('title','count'))
```

```python
data_agg
```

```python
data_agg = data_agg.reset_index()
data_agg
```

```python
data_agg
```

```python
# how to remove multilevel index like if we run set_index two times
# join function
```

```python
data_agg = data_agg.reset_index()
```

```python
data_agg
```

```python
data_agg['year_diff'] = data_agg['year_max']-data_agg['year_min']
```

```python
data_agg['movies_per_year'] = data_agg['title_count']/data_agg['year_diff']
```
